```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("train.csv")
data
```

```python
data1=data[['MSSubClass','LotArea','LotShape','LotConfig','BldgType','HouseStyle','OverallQual','OverallCond','YearBuilt','YearRemodAdd','1stFlrSF','GrLivArea',
            'BedroomAbvGr','YrSold','SaleCondition','SalePrice']]
data1.isna().sum()
```

```python
data1
```

```python
xtrain=data1.iloc[:,:-1].values
ytrain=data1.iloc[:,-1:].values
xtrain
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
xtrain[:,2]=l.fit_transform(xtrain[:,2])
xtrain[:,3]=l.fit_transform(xtrain[:,3])
xtrain[:,4]=l.fit_transform(xtrain[:,4])
xtrain[:,5]=l.fit_transform(xtrain[:,5])
xtrain[:,14]=l.fit_transform(xtrain[:,14])
```

```python
from sklearn.ensemble import RandomForestRegressor
r1= RandomForestRegressor()
r1.fit(xtrain,ytrain)
```

```python
data2=pd.read_csv("test.csv")
data2
```

```python
ids=data2.iloc[:,0].values
data2
```

```python
data5=data2[['MSSubClass','LotArea','LotShape','LotConfig','BldgType','HouseStyle','OverallQual','OverallCond','YearBuilt','YearRemodAdd','1stFlrSF','GrLivArea',
            'BedroomAbvGr','YrSold','SaleCondition']]
data5.isna().sum()
xtest=data5.iloc[:,:].values
```

```python
from sklearn.preprocessing import LabelEncoder
l1=LabelEncoder()
xtest[:,2]=l1.fit_transform(xtest[:,2])
xtest[:,3]=l1.fit_transform(xtest[:,3])
xtest[:,4]=l1.fit_transform(xtest[:,4])
xtest[:,5]=l1.fit_transform(xtest[:,5])
xtest[:,14]=l1.fit_transform(xtest[:,14])
```

```python
ypred=r1.predict(xtest)
ypred
a2=list(ids)
ypred=list(ypred)
```

```python
import csv
for i in range(len(xtest)):
    a=[]
    a.append(int(a2[i]))
    a.append(float(ypred[i]))
    with open('sample_submission.csv','a',newline='') as csv1:
        writer = csv.writer(csv1)
        writer.writerow(a)
        csv1.close()
    
```
